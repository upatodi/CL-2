Write a program for sending sensor data to the cloud and storing it in a database

```python
import http.client, urllib.request, urllib.parse
key=''
 
t=int(22)
while True:
    t=t+1
    print("temperature",t)
    #set parameters
    params = urllib.parse.urlencode({'field':t,'key':key})
    #set headers
    headers={"Content-typeZZe":"application/x-www-form-urlencoded","Accept":"text/plain"}
    #establish a connection
    con=http.client.HTTPConnection("api.thingspeak.com")
    #send request
    con.request("POST","/update",params,headers)
    #get response
    response=con.getresponse()
    #print response status and reason
    print(response.status,response.reason)
    #read the response
    data = response.read()
    con.close()
```
